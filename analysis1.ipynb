# education investigation

```python
def high_education(i):
    if 'doctor' in i.lower():
        return 3

    elif 'phd' in i.lower():
        return 3
    elif 'ph.d.' in i.lower():
        return 3
    elif 'master' in i.lower():
        return 2
    elif 'm.s.' in i.lower():
        return 2
    elif 'm.a.' in i.lower():
        return 2
    elif 'MA' in i:
        return 2
    elif 'MS' in i:
        return 2
    else:
        return 1

# high_education(['PhD, Earth and Environmental Sciences', 'Master of Science - MS, Environmental Science', 'Bachelor of Science (B.S.), Physics'])
# high_education(['Master of Arts (M.A.), Geology/Earth Science', 'Bachelor of Science, Computer Science'])
# high_education(['Juris Doctor, Law', 'Masters, Chemical Engineering', 'Bachelors, Chemical Engineering'])
print(high_education("['Doctor of Education - EdD, Health and Behavioral Science', 'Master, Public Health', 'Advance Certificate in Project Management, Project Management']"))
```

```python
df1['high_ed']=df1['deg'].apply(high_education)
df1.head()
```

```python
df1.groupby(['high_ed']).aggregate("count")
```

```python
df2['high ed']=df2['deg'].apply(high_education)
df2.head()
```

```python
df2.groupby(['high ed']).aggregate("count")
```

```python
def clean_places(work):
    if len(work)==2 or work=='["\'\'"]':
        return np.nan
    else:
        if ',' in work:
            return work[1:work.find(',')]
        else:
            return work[1:-2]
        
        
df1['recent place']=df1['places_work'].apply(clean_places)
df1_filter=df1[df1['recent place'].notnull()]
df1_filter

nyc=['Greater New York City Area','Long Island','New York','Brooklyn','Bronx',"Queens",'Manhattan',
     'Dobbs Ferry','Tarrytown','New York City Metropolitan Area','Melville','New Haven','Edison','Connecticut',
     'Copiague','Hauppauge','Basking Ridge','New Rochelle','Islandia',
     'Greater New York City Area']
#New York, NJ, Connecticut
count1=0
for i in list(df1_filter['recent place']):
    place=i.translate(str.maketrans('', '', string.punctuation))
    if place in nyc:
        print(place)
        count1+=1
print(count1)
print(len(df1_filter)) 
```

```python
df2['recent place']=df2["places_work"].apply(clean_places)
df2_filter=df2[df2['recent place'].notnull()]
count2=0
for i in list(df2_filter['recent place']):
    place=i.translate(str.maketrans('', '', string.punctuation))
    if place in nyc:
        count2+=1
print(count2)
print(len(df2_filter))
```
