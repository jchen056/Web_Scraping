```python
def extract_str_info(A):
    A=A[1:-1]
    B=A.split("', '")
    return B

temp=extract_str_info(df1['yrs_working'][43])
temp
```

```python
def has_numbers(inputString):
    return any(char.isdigit() for char in inputString)

def extract_dates(A):
    B=[]
    for i in A:
        if has_numbers(i):
            B.append(i)
    return B

temp2=extract_dates(temp)
temp2
```

```python
def only_letters(string):
    return all(letter.isalpha() for letter in string)

def extract_places(A,B):
    C=[]
    for i in A:
        if i not in B:
            C.append(i)
    return C
extract_places(temp,temp2)
```
